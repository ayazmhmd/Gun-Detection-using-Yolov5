```python
img_path=os.path.join(os.path.join(path,'dataset'),'images')
lab_path=os.path.join(os.path.join(path,'dataset'),'labels')
i=0
for img in os.listdir(img_path):
  name=img.split('.')
  label=name[0]+".txt"
  image_path=os.path.join(img_path,img)
  label_path=os.path.join(lab_path,label)
  read_img=cv2.imread(image_path)
  read_label=open(label_path, 'r')
  Lines = read_label.readlines()
  
  count = 0
  data=[]
  # Strips the newline character
  for line in Lines:
    if count==0:
      count += 1
      continue
    lis=line.split(' ')
    x1,y1,x2,y2=lis[0],lis[1],lis[2],lis[3]
    x1=int(x1)
    x2=int(x2)
    y1=int(y1)
    y2=int(y2)
    w=x2-x1
    h=y2-y1
    bbx=x1,y1,w,h
    data.append(bbx)
  for bobx in data:
    x1,y1,w,h=bobx
    new_img=Chitra(read_img,[x1, y1,x1+w,y1+h],'')
    new_img.resize_image_with_bbox((128,128))
    bbox=new_img.bboxes
    x1,y1=bbox[0][0]
    x2,y2=bbox[0][1]
    x1=int(x1)
    x2=int(x2)
    y1=int(y1)
    y2=int(y2)
    w=x2-x1
    h=y2-y1
    Xmin = (x1+w/2)/128
    Ymin = (y1+h/2)/128
    W = w/128
    H = h/128
    lab=img.split('.')[0]
    new_im=np.asarray(new_img.image)
    cv2.imwrite(f'data/{img}',new_im)
    with open(f'data/{label}', 'w') as f:
      f.write(str(0))
      f.write(' ')
      f.write(str(Xmin))
      f.write(' ')
      f.write(str(Ymin))
      f.write(' ')
      f.write(str(W))
      f.write(' ')
      f.write(str(H))
      f.close()
```
